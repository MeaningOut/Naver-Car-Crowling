```python
from bs4 import BeautifulSoup
import urllib.request as req

def get_car_list(car_url):
    res = req.urlopen(car_url)
    soup = BeautifulSoup(res, 'html.parser')

    value = []
    tmp = soup.select('.model_name')
    for t in tmp:
        value.append(t.get_text())
        
    return value
```

```python
get_car_list('https://auto.naver.com/car/carKindType.nhn?carKndCd=5&carKndTpCatCd=OWD&importYn=N')
```

```python
from selenium import webdriver

driver = webdriver.Chrome("C:/Users/jeewo/Downloads/chromedriver_win32/chromedriver.exe")
driver.get(import_ven_link)

while True:
    print(driver.current_url)
    mpv_list += get_car_list(driver.current_url)

    # 다음 버튼 클릭
    if driver.find_elements_by_css_selector('#content > div.paginate2 > a.next'):
        driver.find_element_by_css_selector('#content > div.paginate2 > a.next').click()
    else:
        break

driver.close()

mpv_list
```
