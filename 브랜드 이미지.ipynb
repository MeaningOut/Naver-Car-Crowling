```python
from selenium import webdriver
from bs4 import BeautifulSoup as soups
import os

  
def search_selenium(search_name, search_limit, title):
    browser = webdriver.Chrome("C:/Users/jeewo/Downloads/chromedriver_win32/chromedriver.exe")
    search_url = "https://www.google.com/search?q=" + str(search_name) + "&hl=ko&tbm=isch"
    
    browser.get(search_url)
    
    image_count = len(browser.find_elements_by_tag_name("img"))
    
    print("로드된 이미지 개수 : ", image_count)

    browser.implicitly_wait(2)

    for i in range(search_limit) :
        image = browser.find_element_by_id('islmp').find_elements_by_tag_name("img")[i]
        image.screenshot("brand_images/" + title + ".png")
    browser.close()
```

```python
for c in my_list:
    print(c, ' 로고')
    search_selenium(c + ' 로고', 1, c)
```

```python
from bs4 import BeautifulSoup
import urllib.request as req

driver = webdriver.Chrome("C:/Users/jeewo/Downloads/chromedriver_win32/chromedriver.exe")

car_data = []

driver.get('https://auto.naver.com/car/mainList.nhn?importYn=N')
res = req.urlopen(driver.current_url)
soup = BeautifulSoup(res, 'html.parser')

tmp = soup.select("#_vendor_select_layer > div > div.maker_group > div > ul > li > a > img")
for t in tmp:
    print(t.attrs['alt'], ' - ', t.attrs['data-src'])
    car_data.append({"company":t.attrs['alt'], "src":t.attrs['data-src']})

driver.close()
```
